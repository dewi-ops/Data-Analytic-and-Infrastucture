```python
df = pd.read_csv("data/raw/superstore_raw.csv")
df.isnull().sum()
df.duplicated().sum()
```

```python
df = df.dropna(subset=["Sales", "Order Date"])
df = df.drop_duplicates()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
```

```python
df["Profit_Margin"] = df["Profit"] / df["Sales"]
```

```python
df.to_csv(
 "data/clean/superstore_clean.csv",
 index=False
)
```

```python
df_curated = df.copy()
df_curated.to_csv("data/curated/superstore_curated.csv", index=False)
```
